## Задача 1
Бросаем 6-гранный кубик . А - вероятность выпадение 1 или 2. В = отрицание A. Найдите величины A и B. Какое у нас распределение дискретной случайной величины X? 


```python
import numpy as np

# Вероятность события A = 2/6 (1/3), вероятность B соответственно = 2/3
P_A = 1/3
P_B = 2/3

# Распределение случайной дискретной величины X
# По определению мат ожидания имеем теоретическое мат ожидание X:
E_X_th = 1*P_A + 0*P_B

# Возможные выпавшие грани кубика
cases = [1, 2, 3, 4, 5, 6]
# Количество бросков кубика
size = [500]

# Выполняем 500 бросков кубика и отбираем те, где выполнилось событие A
# Распределение у выпадения разных граней кубика равномерное, p не указываем
res = np.random.choice(cases, size=size, p=None)
case_A = np.where(np.isin(res, [1, 2]), 1, 0)

# Посчитаем эмпирическое мат ожидание X
E_X_emp = np.mean(case_A)

print(f"Теоретическая вероятность события A = {P_A}; события B = {P_B}")
print(f"Теоретическое мат ожидание X = {E_X_th}; Эмпирическое мат ожидание X = {E_X_emp}")
```

## Задача 2
Бросается два 6-гранных кубика. Какая вероятность, что на обоих кубиках выпадет нечетное число?

```python
# Вероятность выпадения нечетного числа на одном кубике равна 1/2. 
# Теоретическая вероятность того, что нечетное число выпало на обоих кубиках, равна
P_th = 1/2 * 1/2

# Аналогично первой задаче
cases = [1, 2, 3, 4, 5, 6]
size = [500]

# Выполним броски двух кубиков и отберем те, где на каждом выпало нечетное число
first = np.random.choice(cases, size=size, p=None)
second = np.random.choice(cases, size=size, p=None)
both_odd = (first % 2 == 1) & (second % 2 == 1)

# Посчитаем эмпирическую вероятность
P_emp = np.mean(both_odd)
print(f"Теоретическая вероятность = {P_th}; Эмпирическая вероятность = {P_emp}")
```

## Задача 3
Бросается два 6-гранных кубика, найти вероятность, что произошло и А и В:
А - выпадение нечетного числа на кубике 1 (неявно, что на втором кубике может выпасть все что угодно).
B - выпадение 5-ки на кубике 2 (неявно, что на первом кубике может выпасть все что угодно).

```python
# Необходимо, чтобы на первом кубике выпало нечетное число, а на втором обязательно выпала цифра 5
# Вероятность выпадения нечетного числа на первом кубике: A = 1/2, выпадение числа 2 на втором: B = 1/6
# Теоретическая вероятность события A и B
P_AB_th = 1/2 * 1/6

#Аналогично первой задаче
cases = [1, 2, 3, 4, 5, 6]
size = [500]

# Выполним броски двух кубиков и отберем те результаты, при которых выполняется и A и B
first = np.random.choice(cases, size=size, p=None)
second = np.random.choice(cases, size=size, p=None)
A_and_B = (first % 2 == 1) & (second == 5)

#Посчитаем Эмпирическую вероятность
P_AB_emp = np.mean(A_and_B)
print(f"Теоретическая вероятность = {P_AB_th}; Эмпирическая вероятность = {P_AB_emp}")
```
