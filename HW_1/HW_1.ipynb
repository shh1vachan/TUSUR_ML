## Подключаем библиотеку Pandas и загружаем таблицу из задания:


```python
import pandas as pd
dataframe = pd.read_csv(r"X:\TUSUR_LEARNING\HW_1\data\товары_таблица.csv")
```

## Создаем новую колонку "Цена со скидкой"

```python
dataframe['Цена со скидкой'] = dataframe['Цена']* (100 - dataframe['Скидка в процентах']) / 100
```

```python
dataframe.head()
```

## Сортируем товары по убыванию ценны со скидкой и выводим топ-10 самых дорогих

```python
dataframe = dataframe.sort_values(by = 'Цена со скидкой', ascending = False)
dataframe.head(10)
```

## Группируем данные по категориям и вычисляем среднюю цену товара

```python
grouped = dataframe.groupby('Категория')[['Цена', 'Цена со скидкой']].mean()
grouped
```
